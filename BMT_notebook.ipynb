## Getting the i3d and vggish features from the video

```python
%cd ./submodules/video_features
```

Change the value in video_paths to your video in the next 2 cells

```python
%%bash
source activate i3d
conda info --envs
python main.py \
    --feature_type i3d \
    --on_extraction save_numpy \
    --device_ids 0 \
    --extraction_fps 25 \
    --video_paths /content/BMT/sample/women_long_jump.mp4 \
    --output_path /content/BMT/test 
```

```python
%%bash
source activate vggish
conda info --envs
wget https://storage.googleapis.com/audioset/vggish_model.ckpt -P ./models/vggish/checkpoints
python main.py \
    --feature_type vggish \
    --on_extraction save_numpy \
    --device_ids 0 \
    --video_paths /content/BMT/sample/women_long_jump.mp4 \
    --output_path /content/BMT/test
```

## Running the video prediction file

The output will contain the captions in the video

In the test directory, copy the path of _rbg.npy, _vggish.npy, _flow.npy files generated of your video and paste them to their respective flags in the below cell

```python
%%bash
source activate bmt
python ./sample/single_video_prediction.py \
    --prop_generator_model_path /content/BMT/sample/best_prop_model.pt \
    --pretrained_cap_model_path /content/BMT/best_cap_model.pt  \
    --vggish_features_path /content/BMT/test/women_long_jump_vggish.npy \
    --rgb_features_path /content/BMT/test/women_long_jump_rgb.npy \
    --flow_features_path /content/BMT/test/women_long_jump_flow.npy \
    --duration_in_secs 99 \
    --device_id 0 \
    --max_prop_per_vid 100 \
    --nms_tiou_thresh 0.4
```
